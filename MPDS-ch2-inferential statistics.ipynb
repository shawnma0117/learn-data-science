```python
from scipy.stats import binom
import matplotlib.pyplot as plt
```

#  From binomial to normal distribution

```python
fig, ax = plt.subplots(1,1)
x = range(7)
n, p = 6, 0.5
```

```python
rv = binom(n, p)
```

```python
ax.vlines(x, 0 , rv.pmf(x), colors='k', linestyles='-', lw=1, label='Probablity')
ax.legend(loc = 'best', frameon = False)
plt.show()
```

```python
fig, ax = plt.subplots(2,2,figsize=(15,10))
x = range(101)
n, p = 100, 0.5
rv = binom(n, p)
ax[0,0].vlines(x, 0 , rv.pmf(x), colors='k', linestyles='-', lw=1, label='Probablity')
ax[0,0].legend(loc = 'best', frameon = False)
ax[0,0].set(xlabel='No.of instances',ylabel='Probablity', title = 'when p=0.5')

p=0.4
rv = binom(n, p)
ax[0,1].vlines(x, 0 , rv.pmf(x), colors='k', linestyles='-', lw=1, label='Probablity')
ax[0,1].legend(loc = 'best', frameon = False)
ax[0,1].set(xlabel='No.of instances',ylabel='Probablity', title = 'when p=0.4')

p=0.6
rv = binom(n, p)
ax[1,0].vlines(x, 0 , rv.pmf(x), colors='k', linestyles='-', lw=1, label='Probablity')
ax[1,0].legend(loc = 'best', frameon = False)
ax[1,0].set(xlabel='No.of instances',ylabel='Probablity', title = 'when p=0.6')
plt.show()
```
