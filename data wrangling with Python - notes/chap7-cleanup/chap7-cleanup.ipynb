```python
import csv
```

```python
# 升级版,最终版 code
data_rdr = csv.reader(open('mn.csv', 'rb'))
header_rdr = csv.reader(open('mn_headers.csv', 'rb'))

data_rows = [d for d in data_rdr]
header_rows = [h for h in header_rdr if h[0] in data_rows[0]]

all_short_headers = [h[0] for h in header_rows]

skip_index = [] # 储存在all short headers中找不到的标题的索引
final_header_rows= []

for header in data_rows[0]:
    if header not in all_short_headers:
        index=data_rows[0].index(header)
        skip_index.append(index)
    else:
        for head in header_rows:
            if head[0] == header:
                final_header_rows.append(head)
                break

# print len(skip_index),skip_index

# print final_header_rows[:10]

new_data = [] # 跳过skip_index的列，存放到新的数据

for row in data_rows[1:]: #只取数据部分
    new_row=[]
    for i,d in enumerate(row):
        if i not in skip_index:
            new_row.append(d)
    new_data.append(new_row)  

zipped_data = []

for drow in new_data:
    zipped_data.append(zip(final_header_rows,drow)) #将header与数据行两两匹配起来。

zipped_data[0]
```

### 统计每个问题回答为na的个数 

```python
na_count = {}

for row in zipped_data:
    for resp in row:
        question=resp[0][1]
        anser=resp[1]
        if anser == 'NA':
            if question in na_count.keys():
                na_count[question]+=1
            else:
                na_count[question]=1
```

### 对每个问题统计其不同数据类型的答案的数量

```python
datatypes = {}
start_dict = {'digit':0,'boolean':0,'empty':0,'time_related':0,'text':0,'unknown':0}
```

```python
for row in zipped_data:
    for resp in row:
        question=resp[0][1]
        answer=resp[1]
        key='unknown'
        if answer.isdigit():
            key='digit'
        elif answer in ['Yes','No','True','False']:
            key = 'boolean'
        elif answer.isspace():
            key = 'empty'
        elif answer.isalpha():
            key = 'text'
        elif answer.find('/')>0 or answer.find(':')>0:
            key = 'time_related'
        if question not in datatypes.keys():
            datatypes[question]=start_dict.copy()
        datatypes[question][key] +=1
```

```python
datatypes
```

```python
time_related_sum=0
for type_sum in datatypes.values():
    time_related_sum += type_sum['time_related']
```
